```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("gp-march.csv")
```

```python
sales_ids = []
```

```python
def data_cleaner(data):
    # print(data)
    if data['Sale Id'] in sales_ids:
        # remove the sales amount
        data['Sale Total'] = 0
        return data
    else:
        sales_ids.append(data['Sale Id'])
        return data
```

```python
df1 = df.apply(data_cleaner,axis=1)
```

```python
df1.to_csv('gp-cleaned.csv')
```
